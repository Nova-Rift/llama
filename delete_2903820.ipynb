```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA

# Dataset: differences for numbers from 0 to 9 (10 numbers)
data = [(i, j, i-j) for i in range(10) for j in range(10)]

class SimpleModel(nn.Module):
    def __init__(self, embedding_dim):
        super(SimpleModel, self).__init__()
        self.token_embedding = nn.Embedding(10, embedding_dim)

    def forward(self, x, y):
        x_embed = self.token_embedding(x)
        y_embed = self.token_embedding(y)
        diff = x_embed - y_embed
        print(f'diff = {diff.sum(dim=1)}')
        return diff.sum(dim=1)

# Function to visualize embeddings using PCA
def visualize_embeddings(embeddings, title):
    pca = PCA(n_components=2)
    reduced = pca.fit_transform(embeddings.detach().numpy())
    plt.scatter(reduced[:, 0], reduced[:, 1], c=np.arange(10), cmap='rainbow', marker='o')
    for i, (x, y) in enumerate(reduced):
        plt.annotate(str(i), (x, y))
    plt.title(title)
    plt.colorbar()
    plt.show()

# Hyperparameters
learning_rate = 0.001
embedding_dim = 5
epochs = 1000

# Model, Loss, and Optimizer
model = SimpleModel(embedding_dim)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Visualize initial embeddings
visualize_embeddings(model.token_embedding.weight.data, "Initial Embeddings")


# Test
x_test, y_test = 7, 3
output = model(torch.tensor([x_test]), torch.tensor([y_test]))
print(f"Prediction for the difference between {x_test} and {y_test} is: {output.item():.2f}. Actual difference is: {x_test - y_test}")

# Training Loop
for epoch in range(epochs):
    total_loss = 0
    for x, y, diff in data:
        # Convert to tensors
        x_tensor = torch.tensor([x], dtype=torch.long)
        y_tensor = torch.tensor([y], dtype=torch.long)
        diff_tensor = torch.tensor([diff], dtype=torch.float32)
        
        # Zero the gradients
        optimizer.zero_grad()
        
        # Forward pass
        output = model(x_tensor, y_tensor)
        
        # Calculate loss
        loss = criterion(output, diff_tensor)
        
        # Backward pass and optimization
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
    
    # Print loss every 100 epochs
    if (epoch+1) % 100 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {total_loss/len(data):.4f}')

# Visualize final embeddings
visualize_embeddings(model.token_embedding.weight.data, "Final Embeddings")

# Test
x_test, y_test = 3, 7
output = model(torch.tensor([x_test]), torch.tensor([y_test]))
print(f"Prediction for the difference between {x_test} and {y_test} is: {output.item():.2f}. Actual difference is: {x_test - y_test}")

# Test
x_test, y_test = 7, 3
output = model(torch.tensor([x_test]), torch.tensor([y_test]))
print(f"Prediction for the difference between {x_test} and {y_test} is: {output.item():.2f}. Actual difference is: {x_test - y_test}")
```
